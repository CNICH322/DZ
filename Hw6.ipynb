```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
```

```python
data_2d = pd.read_csv("data_2d.csv")
```

```python
print(data_2d.shape)
print(data_2d.head())
```

```python
kv = []
for k in range(1, 11):
    model = KMeans(n_clusters=k,random_state=0,n_init="auto")
    model.fit(data_2d)
    kv.append(model.inertia_)
```

```python
plt.plot(range(1,11), kv, marker="o")
plt.show()
```

```python
kmeans = KMeans(n_clusters=2,random_state=0,n_init="auto")
labels = kmeans.fit_predict(data_2d)
centers = kmeans.cluster_centers_
```

```python
plt.scatter(data_2d.iloc[:,0],data_2d.iloc[:,1],c=labels)
plt.scatter(centers[:,0],centers[:,1],color="red",marker="X")
plt.show()
```

```python
mnist = pd.read_csv("mnist.csv")
mnist.head()
```

```python
kv = []
for k in range(1,11):
    model = KMeans(n_clusters=k,random_state=0,n_init="auto")
    model.fit(mnist)
    kv.append(model.inertia_)
```

```python
plt.plot(range(1,11), kv, marker="o")
plt.grid()
plt.show()
```

```python
kmeans = KMeans(n_clusters=4,random_state=0,n_init="auto")
labels = kmeans.fit_predict(mnist)
```

```python
pca = PCA(n_components=2)
mnist_2d = pca.fit_transform(mnist)
print(mnist_2d.shape)
```

```python
plt.scatter(mnist_2d[:,0],mnist_2d[:,1],c=labels)
plt.xlabel("PCA 1")
plt.ylabel("PCA 2")

plt.show()
```
